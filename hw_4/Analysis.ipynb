```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# VI and PI Results

```python
gammas = [0.1, 0.5, 0.9]
envs = ['sm', 'lg']
solvers = ['VI', 'PI']
times = []
steps = []

for solver in solvers:
    for gamma in gammas:
        env_times = []
        env_steps = []
        for env in envs:
            gamma_times = []
            gamma_steps = []
            files = glob.glob('output/{}/*{}*{}.csv'.format(solver, env, '_' + str(gamma)))
            for file in files:
                print(file)
                df = pd.read_csv(file, keep_default_na=False)
                df.replace('', np.nan, inplace=True)
                df.dropna(inplace=True)
                df['time'] = df['time'].astype('float')
                df['steps'] = df['steps'].astype('int')
                time = df['time'].sum()
                step = df['steps'].max()
                gamma_times.append(time)
                gamma_steps.append(step)
                print('Solver: {}, Env: {}, Gamma: {}'.format(solver, env, gamma))
                print('Steps: {}, Time: {}'.format(step, time))
                print('-'*20)
            env_times.append(gamma_times)
            env_steps.append(gamma_steps)
        times.append(env_times)
        steps.append(env_steps)
```

## Gammma=0.1, Gammma=0.5, Gammma=0.9

# QL Results

```python
gammas = [0.1, 0.5, 0.9]
envs = ['sm', 'lg']
ql_times = []
ql_steps = []

for gamma in gammas:
    env_times = []
    env_steps = []
    for env in envs:
        gamma_times = []
        gamma_steps = []
        files = glob.glob('output/QL/*{}*zeros_{}*csv'.format(env, gamma))
        for file in files:
            print(file)
            df = pd.read_csv(file, keep_default_na=False)
            df.replace('', np.nan, inplace=True)
            df.dropna(inplace=True)
            df['time'] = df['time'].astype('float')
            df['steps'] = df['steps'].astype('int')
            time = df['time'].sum()
            step = df['steps'].max()
            gamma_times.append(time)
            gamma_steps.append(step)
            print('Solver: {}, Env: {}, Gamma: {}'.format(solver, env, gamma))
            print('Steps: {}, Time: {}'.format(step, time))
            print('-'*20)
        env_times.append(gamma_times)
        env_steps.append(gamma_steps)
    ql_times.append(env_times)
    ql_steps.append(env_steps)
```
