```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.metrics import accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('bank/bank-additional-full.csv', sep=';')
```

```python
df.head()
```

```python
x = df.drop('eye', axis=1)
y = df[['eye']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, shuffle=True, random_state=24)
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred_dt = dt.predict(x_test)
```

```python
accuracy_score(y_pred_dt, y_test)
```

```python
svm = SVC(gamma='scale')
```

```python
svm.fit(x_train, y_train.values.reshape(-1))
```

```python
y_pred_svm = svm.predict(x_test)
```

```python
accuracy_score(y_pred_svm, y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train.values.reshape(-1))
```

```python
y_pred_knn = knn.predict(x_test)
```

```python
accuracy_score(y_pred_knn, y_test)
```
